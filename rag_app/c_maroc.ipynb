```python
import pandas as pd
from pprint import pprint
from datasets import load_dataset
from sentence_transformers import SentenceTransformer
```

```python
ds = load_dataset(
    "atlasia/Atlaset",
    split="train",
    streaming=True # we do not download this dataset we create object refrence o dataset
)
```

```python
# Get first 1000 rows
rows = []
number_of_lines = 10000
for i, row in enumerate(ds):
    rows.append(row)

    if i >= number_of_lines:
        break

print(rows[0])
print(len(rows))
```

```python
documents = []
for row in rows:
    documents.append({"text": row["text"]})
pprint(documents)
```

```python
model = SentenceTransformer("sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2")
```

```python
# Quick sanity check
sample_sentence = documents[0]["text"]
sample_embedding = model.encode(sample_sentence)
print(f"Embedding shape: {sample_embedding.shape}")
```
